```python
import numpy as np
import pandas as pd
import pickle as pkl

import matplotlib.pyplot as plt
```

```python
feature_matrix=pkl.load(open('feature_matrix.pkl', 'rb'))
expert_path=pkl.load(open("expert_path.pkl", 'rb'))
```

```python
alpha=[]
for i in range(108):
    alpha.append(pkl.load(open('alpha_%d.pkl'%(i), 'rb')))
```

```python
from irl.validate import draw_airplane_reward
```

```python
for i, a in enumerate((alpha)):
    if i%40==0:
        r = feature_matrix.dot(a).reshape((1600,))
        draw_airplane_reward("alpha_%d"%(i), r, False)
```

```python
def draw_alpha(_alpha, i):
    r=feature_matrix.dot(_alpha).reshape((1600,))
    draw_airplane_reward("alpha %d"%(i), r, False)

draw_alpha(alpha[0], 0)
draw_alpha(alpha[64], 64)
draw_alpha(alpha[107], 107)
```
